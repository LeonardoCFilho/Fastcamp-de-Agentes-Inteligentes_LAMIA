```python
import requests
import pandas as pd
import time
import logging
```

```python
# Set up logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
```

```python
def fetch_top_steam_games(pages=10):
    """
    Fetches the top video games from Steam using the SteamSpy API.
    Since SteamSpy API returns all games per page in an object, we iterate through pages.
    """
    base_url = "https://steamspy.com/api.php"
    all_games_data = []

    logging.info(f"Starting to fetch top Steam games from SteamSpy API ({pages} pages)...")

    for page in range(0, pages):
        try:
            logging.info(f"Fetching page {page}...")
            # Using the 'all' request with page parameter
            response = requests.get(base_url, params={'request': 'all', 'page': page})
            response.raise_for_status()

            data = response.json()

            if not data:
                logging.info(f"No more data found on page {page}. Stopping.")
                break

            for app_id, game_info in data.items():
                game_record = {
                    'appid': game_info.get('appid'),
                    'name': game_info.get('name'),
                    'developer': game_info.get('developer'),
                    'publisher': game_info.get('publisher'),
                    'score_rank': game_info.get('score_rank'),
                    'positive': game_info.get('positive'),
                    'negative': game_info.get('negative'),
                    'userscore': game_info.get('userscore'),
                    'owners': game_info.get('owners'),
                    'average_forever': game_info.get('average_forever'),
                    'average_2weeks': game_info.get('average_2weeks'),
                    'median_forever': game_info.get('median_forever'),
                    'median_2weeks': game_info.get('median_2weeks'),
                    'price': game_info.get('price'),
                    'initialprice': game_info.get('initialprice'),
                    'discount': game_info.get('discount'),
                    'ccu': game_info.get('ccu') # concurrent users
                }
                all_games_data.append(game_record)

            # Rate limiting delay
            time.sleep(1)

        except requests.exceptions.RequestException as e:
            logging.error(f"Error fetching page {page}: {e}")
            break

    df = pd.DataFrame(all_games_data)
    return df
```

```python
# Editado para entrar no padrão da documentação
df = fetch_top_steam_games(pages=1)

for _, row in df.iterrows():
    # Preço está concatenado
    print(f'("{row["name"]}", "{row["average_2weeks"]}", "{float(row["price"])/100.0}", "{row["owners"]}"),')
```
